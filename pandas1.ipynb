```python
import numpy as np 
```

```python
array = np.array([10,20,30,40,50,60,70,80,90])
array
```

```python
for a,b in enumerate (array):      # enumerate() function is used to get the index and value of the array
    print(f"array[{a}] : {b}")
```

```python
array[0] = 100
array
```

```python
print(array[3:6])    # slicing the array from index 3 to 5
```

```python
array[1:3] = 101,102
array
```

```python
arr =np.append(array,110)       # adding at the end of the array
print("arr:",arr,end="\n\n")

arr1 = np.insert(arr,0,100)     # adding at the beginning of the array
print("arr1:",arr1)
```
